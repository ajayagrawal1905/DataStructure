```python
# Iterative Function to calculate (x^y) in O(log y)
def power(x, y,M):
	
	# Initialize result
	res = 1
	
	while (y > 0):
	
		# If y is odd, multiply x with result
		if ((y & 1) != 0):
			res = ((res%M) * (x%M))%M

		# y must be even now
		y = y >> 1 # y = y/2
		x = ((x%M) * (x%M))%M # Change x to x^2
	return res

x = 2
y = int(input())
M = 1000000007
answer = power(x,y,M)
print(answer)
```

```python
def bubble_sort(nums):
    # We set swapped to True so the loop looks runs at least once
    swapped = True
    count = 0
    
    while swapped:
        count+=1
        print(count)
        swapped = False
        for i in range(len(nums) - 1):
            
            if nums[i] > nums[i + 1]:
                # Swap the elements
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                # Set the flag to True so we'll loop again
                print(nums)
                swapped = True


# Verify it works
random_list_of_nums = [5, 2, 1, 8, 4]
bubble_sort(random_list_of_nums)
print(random_list_of_nums)
```

```python
def sort(nums):    
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(nums)-2):
            if nums[i] > nums[i+1]:
                nums[i],nums[i+1] = nums[i+1],nums[i]
                swapped = True
                
nums = map(int, input().split())
nums = list(nums)
sort(nums)
%timeit sort(nums)
print(nums)
```

```python
def selection_sort(nums):
    # This value of i corresponds to how many values were sorted
    for i in range(len(nums)):
        # We assume that the first item of the unsorted segment is the smallest
        lowest_value_index = i
        # This loop iterates over the unsorted items
        for j in range(i + 1, len(nums)):
            print(j,lowest_value_index,nums[j] , nums[lowest_value_index])
            
            if nums[j] < nums[lowest_value_index]:
                lowest_value_index = j
        # Swap values of the lowest unsorted element with the first unsorted
        # element
        nums[i], nums[lowest_value_index] = nums[lowest_value_index], nums[i]
        print(nums)
        #print(nums[i], nums[lowest_value_index] ,'|', nums[lowest_value_index], nums[i])

# Verify it works
random_list_of_nums = [6,1,7,8]
selection_sort(random_list_of_nums)
print(random_list_of_nums)
```

```python
def insertion_sort(nums):
    # Start on the second element as we assume the first element is sorted
    for i in range(1, len(nums)):
        item_to_insert = nums[i]
        
        # And keep a reference of the index of the previous element
        j = i - 1
        # Move all items of the sorted segment forward if they are larger than
        # the item to insert
        while j >= 0 and nums[j] > item_to_insert:
            nums[j + 1] = nums[j]
            print(nums[j + 1] , nums[j])
            j -= 1
        # Insert the item
        nums[j + 1] = item_to_insert
        print(nums[j + 1] , item_to_insert)

# Verify it works
random_list_of_nums = [9, 1, 15, 28, 6]
insertion_sort(random_list_of_nums)
print(random_list_of_nums)
```
